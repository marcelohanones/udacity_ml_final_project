```python
def tokenizer_split(text):
    return [porter.stem(w) for w in text.split(' ')]

def tokenizer_porter(text):
    return [porter.stem(w) for w in text.split(' ') if not w in stop_words ]

tfidf = TfidfVectorizer(strip_accents=None, lowercase=False, preprocessor=None)
```

# Teste n_jobs

```python
# testar se o pipe com Truncated da problema com n_jobs=-1
# 200 samples
pipe_9 = Pipeline([
    ('col', Col_Extractor(['review'])),
    ('prep', Preprocessor()),
    ('to_array', To_array()),
    ('vec', CountVectorizer(vocabulary=vocab)),
    ('tf-idf', TfidfTransformer()),
    ('TruncatedSVD', TruncatedSVD(n_components=1000)),
    ('Normalize', Normalizer(copy=False))
])
master.evaluate_CV(pipe_9, ['SGDC'])
```

```python
# retirando o truncated por Kbest e deixando n_jobs=-1
# 200 samples
pipe_9 = Pipeline([
    ('col', Col_Extractor(['review'])),
    ('prep', Preprocessor()),
    ('to_array', To_array()),
    ('vec', CountVectorizer(vocabulary=vocab)),
    ('tf-idf', TfidfTransformer()),
    ('kbest', SelectPercentile(chi2)),
    ('Normalize', Normalizer(copy=False))
])
master.evaluate_CV(pipe_9, ['SGDC'])
```
